**Phytoplankton df2**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from google.colab import files


# -----------------------
# 1. Load and clean data
# -----------------------


# Upload file manually
uploaded = files.upload()  # This will open a file selector

# Get the filename of the uploaded file
filename = list(uploaded.keys())[0]

# Load it with pandas
df = pd.read_csv(filename)
print(df.head())
```

**Sharks df3**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from google.colab import files


# -----------------------
# 1. Load and clean data
# -----------------------


# Upload file manually
uploaded = files.upload()  # This will open a file selector

# Get the filename of the uploaded file
filename = list(uploaded.keys())[0]

# Load it with pandas
df = pd.read_csv(filename)
print(df.head())
```

```python
# -----------------------
# 2. Clean column names (strip spaces, lowercase)
# -----------------------
df.columns = df.columns.str.strip().str.lower()

# -----------------------
# 3. Rename columns
# -----------------------
df3 = df.rename(columns={
    "longitude": "longitude",
    "latitude": "latitude",
    "meansharkspatialdensity": "shark_density"
})

print("Columns after renaming:")
print(df3.columns.tolist())
print(df3.head())

# -----------------------
# 4. Scatter plot of shark density
# -----------------------
plt.figure(figsize=(8,6))
sc = plt.scatter(df3["longitude"], df3["latitude"], c=df3["shark_density"], cmap="viridis", s=10)
plt.colorbar(sc, label="Shark Spatial Density")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geographical Distribution of Shark Density")
plt.show()

# -----------------------
# 5. Histogram of shark density
# -----------------------
plt.figure(figsize=(6,4))
plt.hist(df3["shark_density"], bins=30, color="skyblue", edgecolor="black")
plt.xlabel("Shark Spatial Density")
plt.ylabel("Frequency")
plt.title("Distribution of Shark Density")
plt.show()
```

Joining Dataframes

```python
import pandas as pd
from scipy.spatial import cKDTree
import numpy as np

# Assume:
# df_temp -> contains latitude, longitude, temperature
# df_species -> contains latitude, longitude, prochlorococcus, synechococcus, picoeukaryotes
# df_shark -> contains latitude, longitude, shark_density

# -----------------------
# 1. Build KD-Trees
# -----------------------
coords_temp = df1[['latitude','longitude']].values
coords_species = df2[['latitude','longitude']].values
coords_shark = df3[['latitude','longitude']].values

tree_species = cKDTree(coords_species)
tree_shark = cKDTree(coords_shark)

# -----------------------
# 2. Find nearest neighbors
# -----------------------
dist_species, idx_species = tree_species.query(coords_temp)
dist_shark, idx_shark = tree_shark.query(coords_temp)

# -----------------------
# 3. Build merged dataframe starting from temperature
# -----------------------
df_merged = df1.copy()

# Add species features
df_merged[['prochlorococcus','synechococcus','picoeukaryotes']] = df2.iloc[idx_species][['prochlorococcus','synechococcus','picoeukaryotes']].values

# Add shark density
df_merged['shark_density'] = df3.iloc[idx_shark]['shark_density'].values

# -----------------------
# 4. Verify
# -----------------------
print("Columns in merged dataframe:")
print(df_merged.columns.tolist())
print(df_merged.head())
```

```python
# Check if 'temperature' exists in the merged dataframe
if 'temperature' in df_merged.columns:
    print("Temperature column exists in df_merged.")
else:
    print("Temperature column NOT found in df_merged.")

# Inspect first few rows
print(df_merged.head())
```

**Save filtered dataset**


```python
# Drop latitude and longitude columns
df_merged = df_merged.drop(columns=['latitude', 'longitude'])

# Verify
print(df_merged.head())
print(df_merged.columns.tolist())

# Drop rows that contain any NaN
df_merged_clean = df_merged.dropna()

# Verify
print("Number of rows before cleaning:", len(df_merged))
print("Number of rows after dropping rows with missing data:", len(df_merged_clean))
print(df_merged_clean.head())

```

```python
# Save the cleaned dataframe to CSV
df_merged_clean.to_csv("merged_cleaned_data.csv", index=False)

print("DataFrame saved to merged_cleaned_data.csv")
```
